```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
print(df)
```

```python
# prompt: create a pie chart with count of transmission. label count of 1 as automatic and 0 as manual. add percentages

import matplotlib.pyplot as plt
# Count the number of cars for each transmission type
transmission_counts = df['am'].value_counts()

# Create labels with counts
labels = ['Manual ({})'.format(transmission_counts[1]), 'Automatic ({})'.format(transmission_counts[0])]

# Create a pie chart
plt.pie(transmission_counts, labels=labels, autopct='%1.1f%%')
plt.title('Count of Cars by Transmission Type')
plt.show()
```

```python
# prompt: create a 3 dimensional donut chart with count records by AM. Translate values in the AM variable: 0 = automatic, 1 = manual

import plotly.graph_objects as go

# Translate values in the AM variable
df['am'] = df['am'].replace({0: 'Automatic', 1: 'Manual'})

# Count the number of cars for each transmission type
transmission_counts = df['am'].value_counts()

# Create a donut chart
fig = go.Figure(data=[go.Pie(labels=transmission_counts.index, values=transmission_counts.values, hole=.3)])
fig.update_layout(title='Count of Cars by Transmission Type')
fig.show()
```
